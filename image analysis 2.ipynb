```python
import cv2
import matplotlib.pyplot as plt

image_path = 'short exposure.tif'
image = cv2.imread(image_path)

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

median_filtered_3x3 = cv2.medianBlur(image_rgb, 3)
median_filtered_5x5 = cv2.medianBlur(image_rgb, 5)

plt.figure(figsize=(15, 10))

plt.subplot(1, 3, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')  

plt.subplot(1, 3, 2)
plt.imshow(median_filtered_3x3)
plt.title('3x3 Median Filter')
plt.axis('off')  

plt.subplot(1, 3, 3)
plt.imshow(median_filtered_5x5)
plt.title('5x5 Median Filter')
plt.axis('off')  

plt.show()
```

```python
image_path = 'long exposure.tif'

# Reading images and converting them to greyscale
image = cv2.imread(image_path, 0) 

if image is not None:
    dft = cv2.dft(np.float32(image), flags=cv2.DFT_COMPLEX_OUTPUT)
    dft_shift = np.fft.fftshift(dft)

    magnitude_spectrum = 20 * np.log(cv2.magnitude(dft_shift[:, :, 0], dft_shift[:, :, 1]) + 1)

    plt.figure(figsize=(12, 6))

   
    plt.subplot(1, 2, 1)
    plt.imshow(image, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')  

    plt.subplot(1, 2, 2)
    plt.imshow(magnitude_spectrum, cmap='gray')
    plt.title('Fourier Transform')
    plt.axis('off') 

    plt.show()
else:
    print("Image not found or path is incorrect.")
```
